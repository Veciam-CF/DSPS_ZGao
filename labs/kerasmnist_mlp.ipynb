```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pylab as plt
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Input
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
np.unique(x_train)
```

```python
x_train = x_train.astype(float) / 255.
x_test = x_test.astype(float) / 255.
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
x_train.shape
```

```python
for i in range(10):
  plt.imshow(x_train[i])
  plt.show()
```

```python
y_train[:10]
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
x_trainraveled = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2]*x_train.shape[3])
```

```python
x_testraveled = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]*x_test.shape[3])
```

```python
x_test.shape
```

```python
x_testraveled.shape
```

```python
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]*x_test.shape[3])
```

```python
mlp = keras.Sequential([
    layers.Dense(320, activation='sigmoid'),
    layers.Dense(32, activation='sigmoid'),
    layers.Dense(32, activation='sigmoid'),
    layers.Dense(10, activation='softmax')
])
```

```python
mlp.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = mlp.fit(x_trainraveled, y_train, batch_size=120, epochs=10, validation_split=0.25, verbose=1)
```

```python
history.history
```

```python
plt.plot(history.history['loss'], label="training")
plt.plot(history.history['val_loss'], label="validation")
plt.xlabel("epoch")
plt.ylabel("log_loss")
plt.legend()
plt.yscale('log')
```

```python
mlp.evaluate(x_testraveled, y_test)
```

```python
input_shape = (28,28,1)

model = Sequential([
    Input(shape=input_shape),
    Conv2D(32, (3,3), activation="relu", padding="same"),
    Conv2D(64, (3,3), activation="relu", padding="same"),
    MaxPooling2D((2,2)),
    Conv2D(128, (3,3), activation="relu", padding="same"),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation="relu"),
    Dense(10, activation="softmax")
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
cnn_history = model.fit(x_train, y_train, batch_size=120, epochs=10, validation_split=0.25, verbose=1)
```

```python
x_test = x_test.reshape(-1, 28, 28, 1)
model.evaluate(x_test, y_test)
```

```python
plt.plot(cnn_history.history['loss'], label="training")
plt.plot(cnn_history.history['val_loss'], label="validation")
plt.xlabel("epoch")
plt.ylabel("log_loss")
plt.legend()
plt.yscale('log')
```

We could see, as the epoch increase, the model still learning from training. However, the validation loss keeps the same. Which may suggest that our model start to remember the training set itself, which is overfiting.
